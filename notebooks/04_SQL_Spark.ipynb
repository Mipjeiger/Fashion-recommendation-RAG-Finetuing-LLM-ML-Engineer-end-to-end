## Insert Spark dataframe into a SQL table

```python
from pyspark.sql import SparkSession
import findspark

findspark.init()

spark = (
    SparkSession.builder
    .appName("Fashion-SQL")
    .master("local[*]")
    .config("spark.sql.shuffle.partitions", "8")
    .config("spark.driver.memory", "4g")
    .getOrCreate()
)
```

```python
sql = spark.read.parquet("hdfs/fashion_items_parquet")
sql.createOrReplaceTempView("fashion_items")

spark.sql(
    """
SELECT * FROM fashion_items;
"""
).show()
```
